```python
import os

from groq import Groq
from google.colab import userdata

client = Groq(api_key=userdata.get('api_key'))
```

```python
user_input = input("Say something: ")
print(user_input)
```

```python
personas = {
    "friendly": "You are a friendly assistant who replies sarcastically.",
    "pm_mentor": "You are an expert Product Manager Mentor. You don't let anyone think less of themselves. Be concise and actionable.",
    "strict" : "You are strict and non-nonsense teachnical tutor."
}

print("Choose a persona: friendly / pm_mentor/ strict")
choice = input("Persona: ").strip().lower()
system_prompt = personas.get(choice, personas["friendly"])

def generate_response(user_msg):
  response = client.chat.completions.create(
      model="llama-3.3-70b-versatile",
      messages = [
          {
          "role" : "user", "content": user_msg
      }
      ]
  )
  return response.choices[0].message.content

conversation = [{"role": "system", "content":system_prompt}]

def generate_response_whist(user_msg):
  try:
    conversation.append({"role":"user", "content": user_msg})
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages = conversation )

    bot_msg = response.choices[0].message.content

    conversation.append({"role":"assistant", "content": bot_msg})

    return bot_msg

  except Exception as e:
    return f"Error: {str(e)}"


while True:
  user_input = input("You : ")

  if user_input.lower() in ["exits", "quit"]:
    break

  reply = generate_response_whist(user_input)
  print("Bot: ", reply)
```
